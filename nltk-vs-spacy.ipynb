```python
import nltk
import MeCab
import spacy
from nltk.corpus import stopwords
import unicodedata

nltk.download('stopwords')
```

```python
japanese_text = """
私は毎日ブログを書いています。自然言語処理はとても面白い分野です。
Pythonと機械学習を使って、日本語のテキストを処理する方法を学びましょう。
"""

print(" Input Text:\n", japanese_text)
```

```python
tagger = MeCab.Tagger("-Owakati")
tokens_nltk = tagger.parse(japanese_text).strip().split()

japanese_stopwords = ['は', 'と', 'を', 'に', 'で', 'が', 'の', 'ます', 'です', '毎日', '使って']

filtered_nltk = [token for token in tokens_nltk if token not in japanese_stopwords]

print("Tokens (NLTK + MeCab):", tokens_nltk)
print("\nStopwords Removed (NLTK):", filtered_nltk)
```

```python
nlp = spacy.load("ja_core_news_md")

doc = nlp(japanese_text)
tokens_spacy = [token.text for token in doc]
stopwords_spacy = [token.text for token in doc if token.is_stop]

filtered_spacy = [token.text for token in doc if not token.is_stop]

print("Tokens (SpaCy):", tokens_spacy)
print("\nStopwords (SpaCy):", stopwords_spacy)
print("\nStopwords Removed (SpaCy):", filtered_spacy)
```

```python
print(f"\nTotal Tokens (NLTK): {len(tokens_nltk)} | Stopwords Removed: {len(filtered_nltk)}")
print(f"Total Tokens (SpaCy): {len(tokens_spacy)} | Stopwords Removed: {len(filtered_spacy)}")
```
